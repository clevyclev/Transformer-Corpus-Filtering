```python
safe_followers = set(('ROOT', 'advmod', 'aux', 'auxpass', 'ccomp'))

new_data = []
bad_data = []
for line in data:
    doc = nlp(line)
    parse = [token.dep_ for token in doc]
    for i, pos in enumerate(parse):
        if pos == 'nsubj' or pos == 'nsubjpass':
            if i + 1 < len(parse):
                if parse[i + 1] in safe_followers:
                    new_data.append(line)
                else:
                    bad_data.append(line)
```

```python
from tqdm import tqdm
```

```python
with open('data/full_corpus/train.txt') as f:
    data = [line.rstrip() for line in f.readlines()]
```

```python
safe_followers = set(('ROOT', 'aux', 'auxpass', 'ccomp', 'advcl', 'relcl'))

new_data = []
bad_data = []
with open('data/preprocessed_data/filtered.txt', 'w') as f:
    for line in tqdm(data):
        doc = nlp(line)
        parse = [token.dep_ for token in doc]
        contiguous_SV = True
        for i, pos in enumerate(parse):
            if pos == 'nsubj' or pos == 'nsubjpass':
                if i + 1 < len(parse):
                    if parse[i + 1] not in safe_followers:
                        contiguous_SV = False
        if contiguous_SV:
            new_data.append(line) #Purely for reviewing the data deemed "good"
            f.write(line + '\n')
        else:
            bad_data.append(line) #Purely for reviewing the data deemed "bad"
```
